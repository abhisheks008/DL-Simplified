```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

## loading the dataset

```python
val_root="/kaggle/input/vegetable-image-dataset/Vegetable Images/validation/"
train_root="/kaggle/input/vegetable-image-dataset/Vegetable Images/train/"
test_root="/kaggle/input/vegetable-image-dataset/Vegetable Images/test/"
```

```python
vegetables_train=os.listdir(train_root)
vegetables_train
```

## Creation of dataframe
**sorting and shuffling the created dataframe individually**

### **train images**

```python
images_train=[]
y_train=[]
for vegetable in os.listdir(train_root):
    directory = train_root + vegetable + "/"
    print(directory)
    for image in os.listdir(directory):
        images_train.append(directory + image)
        y_train.append(vegetable)
```

```python
Images_train=pd.Series(images_train, name="Image").astype(str)
Y_train=pd.Series(y_train, name="Y")
```

```python
train=pd.concat([Images_train, Y_train], axis=1)
train
```

```python
train.sort_values("Y", inplace=True)
train
```

```python
train.reset_index(drop=True, inplace=True)
train
```

### **validation images**

```python
images_val=[]
y_val=[]
for vegetable in os.listdir(val_root):
    directory = val_root + vegetable + "/"
    print(directory)
    for image in os.listdir(directory):
        images_val.append(directory + image)
        y_val.append(vegetable)
```

```python
Images_val=pd.Series(images_val, name="Image").astype(str)
Y_val=pd.Series(y_val, name="Y")
```

```python
validate=pd.concat([Images_val, Y_val], axis=1)
validate
```

```python
validate.sort_values("Y", inplace=True)
validate.reset_index(drop=True, inplace=True)
validate
```

 ###  **Testing Images**

```python
images_test=[]
y_test=[]
for vegetable in os.listdir(test_root):
    directory = test_root + vegetable + "/"
    print(directory)
    for image in os.listdir(directory):
        images_test.append(directory + image)
        y_test.append(vegetable)
```

```python
Images_test=pd.Series(images_test, name="Image").astype(str)
Y_test=pd.Series(y_test, name="Y")
```

```python
test=pd.concat([Images_test, Y_test], axis=1)
test
```

```python
test.sort_values("Y", inplace=True)
test.reset_index(drop=True, inplace=True)
test
```

## ImageDataGenerator

```python
dev_generator = image.ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2, 
    zoom_range=0.2,    
)

test_generator = image.ImageDataGenerator(rescale=1./255)
```

## ResNet50V2 Model for Image Classification

The following code demonstrates the implementation of a ResNet50V2 model for image classification using Keras. The ResNet50V2 model is pre-trained on the ImageNet dataset and is used as a feature extractor.

### importing necessary libaries

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, BatchNormalization, GlobalAveragePooling2D
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from keras.optimizers import Adam
from keras.applications import ResNet50V2
```

### Setting up the Model

```python
adam = Adam(learning_rate=0.001)

model = Sequential()
```

```python

# Load pre-trained ResNet50V2 model (excluding the top layer)
resnet = ResNet50V2(include_top=False, weights='imagenet', input_shape=(100, 100, 3))
```

4. **Building the Model Architecture:** 

A sequential model is created. The base EfficientNetB0 model is added as the first layer. Global average pooling is applied to reduce the spatial dimensions of the output. Dropout layers are introduced to mitigate overfitting. Two dense layers with ReLU activation are added, followed by a final dense layer with softmax activation for multi-class classification.



```python

# Freeze the layers in the pre-trained model
for layer in resnet.layers:
    layer.trainable = False

model.add(resnet)
# Add additional layers for classification
model.add(GlobalAveragePooling2D())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(15, activation='softmax'))
```

5. **Compiling the Model:** 

We compile the model by specifying the loss function, optimizer, and evaluation metrics. In this case, we use categorical cross-entropy as the loss function, the Adam optimizer, and track accuracy as evaluation metrics.

```python
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
model.summary()
```

6. **Training the Model:** We fit the model to the training data (`train_images`) and validate it on the validation data (`val_images`) using the provided training parameters after defining `earlystopping` and reduction in learning rate using `reduce_lr`


```python
# Define the early stopping and learning rate reduction callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=10, verbose=1, min_delta=1e-4)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=4, verbose=1, min_delta=1e-4)
callbacks_list = [early_stop, reduce_lr]
```

### Model Evaluation

The model is trained for 30 epochs with early stopping and learning rate reduction callbacks to prevent overfitting. The validation accuracy and loss are monitored during training.

```python
# Train the model
history = model.fit(train_images, validation_data=val_images, epochs=30, callbacks=callbacks_list)
```

### Conclusion

we have implemented a ResNet50V2 model for image classification using Keras. The ResNet50V2 model, pre-trained on the ImageNet dataset, serves as a powerful feature extractor. By freezing the pre-trained layers and adding additional layers for classification, we were able to achieve good performance on our image classification task.
